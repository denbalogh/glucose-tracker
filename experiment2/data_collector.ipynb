```python
import serial
import random
import string
import os
```

```python
arduino = serial.Serial(port='/dev/cu.usbmodem14201', baudrate=115200, timeout=.1)

fasted = "data/batch4/fasted/raw"
after_meal = "data/batch3/after_meal/raw"
```

```python
for i in range(10):
    hash = ''.join(random.choices(string.ascii_uppercase + string.digits, k=10))
    f = open(f'{fasted}/raw_{hash}.txt', 'w')
    for x in range(50):
        arduino.write(b'get')
        while True:
            if arduino.in_waiting > 0:
                line = arduino.readline()
                f.write(line.decode())
                print(f"File: {i}, entry {x} added")
                break
    f.close()
```

```python
current = fasted
all_entries = ""

for name in os.listdir(current):
    if("raw_" in name):
        f = open(f'{current}/{name}', 'r')
        entries = f.read()
        all_entries += entries
        f.close()

f = open(f'{current}/../fasted.csv', 'w')
f.write(all_entries)
f.close()
```
